### Checking for Missing Values

Before cleaning anything, let's see which columns have missing values and how many — separately for train and test.

```python
# This line of code prints all columns where there are missing values in train data set.
train_cols = train_data.columns
missing_train_cols = []
missing_train_val = []
for col in train_cols:
    if train_data[col].isnull().sum() > 0:
        missing_train_cols.append(col)
        missing_train_val.append(train_data[col].isnull().sum())
        print(f"column {col} have {train_data[col].isnull().sum()} missing values.")
# print(missing_train_cols,missing_train_val)
```

```python
# Since PoolQC, MiscFeature have large missing values so dropped it
train_data.drop(['PoolQC','MiscFeature'],axis=1,inplace=True)
```

# Imputation Of Missing Values (Most Imp.)

#### Imputing Train Data

- Numerical columns like `LotFrontage`, `MasVnrArea`, `GarageYrBlt` are filled with **median** since median is robust to outliers.
- `Electrical` had just a couple of missing rows so we used **mode** (most frequent value).
- For basement and garage related columns, missing values actually mean the house doesn't have that feature, so we fill with `"None"`.
- Same logic for columns like `Alley`, `PoolQC`, `Fence` etc. — high missing % because most houses simply don't have them.

```python
# Here I have filled numerical values with median
train_data['LotFrontage'].fillna(train_data['LotFrontage'].median(), inplace=True)
train_data['MasVnrArea'].fillna(train_data['MasVnrArea'].median(), inplace=True)
train_data['GarageYrBlt'].fillna(train_data['GarageYrBlt'].median(), inplace=True)

# Here I have filled categorical values with mode (there were few missing values)
train_data['Electrical'].fillna(train_data['Electrical'].mode()[0], inplace=True)

# Here I have filled categorical values with median
bsmt_cols = ['BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2']
for col in bsmt_cols:
    train_data[col].fillna('None', inplace=True)

# Filled Garage features having missing values with "None" that means "No Garage"
garage_cols = ['GarageType', 'GarageFinish', 'GarageQual', 'GarageCond']
for col in garage_cols:
    train_data[col].fillna('None', inplace=True)

# columns having high missing values replaced with None as well.
high_missing_cols = ['Alley', 'FireplaceQu', 'Fence','PoolQC','MiscFeature']
for col in high_missing_cols:
    train_data[col].fillna('None', inplace=True)

# Special case
train_data['MasVnrType'].fillna('None', inplace=True)

# Check if any missing values remain
print(train_data.isnull().sum().sort_values(ascending=False).head(10))
```

# Next Steps
# 1) Apply One hot encoding/Label encoding on categorical columns Since ML models only takes numerical values as input.

### Label Encoding Categorical Columns

ML models need numerical input, so we convert all categorical (text) columns to numbers using `LabelEncoder`. We fit the encoder on training data and then apply the same mapping to test data. If test data has a category the encoder never saw during training, we replace it with the most frequent training category to avoid errors.

```python
from sklearn.preprocessing import LabelEncoder
import numpy as np
```

```python
categorical_cols = train_data.select_dtypes(include=['object']).columns
```

```python
categorical_cols
```

```python
label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    train_data[col] = le.fit_transform(train_data[col].astype(str))
    label_encoders[col] = le
```

## Part A: Decision Tree Algorithm

We will train a Decision Tree Classifier on the house price data using four different hyperparameter combinations:

| # | Criterion | max_depth |
|---|-----------|----------|
| 1 | entropy   | 4        |
| 2 | entropy   | 5        |
| 3 | gini      | 4        |
| 4 | gini      | 5        |

Since `SalePrice` is a continuous variable and entropy/gini are classification criteria, we first convert `SalePrice` into price categories (Low, Medium, High) based on percentile ranges. This way we can properly use a classifier here.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
import pandas as pd
import numpy as np
```

### Train-Test Split (70:30)

We'll use the 70:30 split for training and evaluating all four Decision Tree models.

```python
X_train = pd.read_csv('/content/clean_train_70.csv', index_col=0)
X_test = pd.read_csv('/content/clean_test_30.csv', index_col=0)
```

```python
X_train.shape
```

### Preparing Features and Target

```python
y_train = X_train['PriceCategory']
y_test = X_test['PriceCategory']

X_train = X_train.drop(['SalePrice', 'PriceCategory'], axis=1)
X_test = X_test.drop(['SalePrice', 'PriceCategory'], axis=1)
```

### Model 1 — Criterion: Entropy, max_depth: 4

```python
dt1 = DecisionTreeClassifier(criterion='entropy', max_depth=4, random_state=42)
dt1.fit(X_train, y_train)
y_pred1 = dt1.predict(X_test)

print(f'Accuracy: {accuracy_score(y_test, y_pred1):.4f}')
```

### Model 2 — Criterion: Entropy, max_depth: 5

```python
dt2 = DecisionTreeClassifier(criterion='entropy', max_depth=5, random_state=42)
dt2.fit(X_train, y_train)
y_pred2 = dt2.predict(X_test)

print(f'Accuracy: {accuracy_score(y_test, y_pred2):.4f}')
```

### Model 3 — Criterion: Gini, max_depth: 4

```python
dt3 = DecisionTreeClassifier(criterion='gini', max_depth=4, random_state=42)
dt3.fit(X_train, y_train)
y_pred3 = dt3.predict(X_test)

print(f'Accuracy: {accuracy_score(y_test, y_pred3):.4f}')
```

### Model 4 — Criterion: Gini, max_depth: 5

```python
dt4 = DecisionTreeClassifier(criterion='gini', max_depth=5, random_state=42)
dt4.fit(X_train, y_train)
y_pred4 = dt4.predict(X_test)

print(f'Accuracy: {accuracy_score(y_test, y_pred4):.4f}')
```

### Hyperparameter Tuning using GridSearchCV

Instead of manually trying each combination of criterion and max_depth like we did above, scikit-learn gives us a built-in way to do this automatically — `GridSearchCV`. It takes a dictionary of hyperparameters, tries every possible combination, evaluates each using cross-validation, and tells us which combination performed the best. Think of it as a smarter way of doing what we just did manually.

```python
from sklearn.model_selection import GridSearchCV

# defining the hyperparameters we want to try
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [4, 5]
}

# setting up grid search with 5-fold cross validation
dt = DecisionTreeClassifier(random_state=42)
grid_search = GridSearchCV(dt, param_grid, cv=10, scoring='accuracy')

# fitting on training data (using the 70:30 split)
grid_search.fit(X_train, y_train)

# results
print("All combinations tried:")
cv_results = pd.DataFrame(grid_search.cv_results_)[['param_criterion', 'param_max_depth', 'mean_test_score']]
cv_results.columns = ['Criterion', 'max_depth', 'Avg CV Accuracy']
print(cv_results.to_string(index=False))

print(f"\nBest Parameters: {grid_search.best_params_}")
print(f"Best CV Accuracy: {grid_search.best_score_:.4f}")
print(f"Test Accuracy: {grid_search.score(X_test, y_test):.4f}")
```
